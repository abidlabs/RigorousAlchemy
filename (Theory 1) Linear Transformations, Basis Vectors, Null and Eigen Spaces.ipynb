# Linear algebra is about transformations

```python
def plot_vectors():
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=[12,4], facecolor="black")
    ax = axes[0]
    ax.axhline(0, color='white')
    ax.axvline(0, color='white')
    ax.set_xlim([-5, 5])
    ax.set_ylim([-5, 5])
    ax.set_title("Rotator")
    ax.arrow(0,0,1,1,width=0.1)
    ax.arrow(0,0,1,-1,width=0.1, color='C1')
    ax.annotate('Rotates a 2D vector clockwise by 90 deg.', (0,0), (0, -25), xycoords='axes fraction', textcoords='offset points', va='top', size=12,);

    ax = axes[1]
    ax.axhline(0, color='white')
    ax.axvline(0, color='white')
    ax.set_xlim([-5, 5])
    ax.set_ylim([-5, 5])
    t = np.linspace(0,np.pi*2,100)
    ax.plot(2.1*np.cos(t), 2.1*np.sin(t), linewidth=1)
    ax.set_title("Unitizer")
    ax.arrow(0,0,2,2,width=0.1)
    ax.arrow(0,0,np.sqrt(2)/1.25,np.sqrt(2)/1.25,width=0.1, color='C1')
    ax.annotate('Scales a 2D vector to have unit length.', (0,0), (0, -25), xycoords='axes fraction', textcoords='offset points', va='top', size=12,);
    
    ax = axes[2]
    ax.axhline(0, color='white')
    ax.axvline(0, color='white')
    ax.set_xlim([-5, 5])
    ax.set_ylim([-5, 5])
    ax.set_title("Shifter")
    ax.arrow(0,0,3,2,width=0.1)
    ax.arrow(0,0,3,3,width=0.1, color='C1')
    ax.annotate('Shifts a 2D vector vertically up by 1 unit.', (0,0), (0, -25), xycoords='axes fraction', textcoords='offset points', va='top', size=12,);
```

```python
plot_vectors()
```
